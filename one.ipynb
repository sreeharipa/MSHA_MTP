```python
special_chars = ['am', 'pm', ':', ' noon']
for char in special_chars:
    df['shift'] = df['shift'].str.replace(char, '')

df['shift'] = df['shift'].str.replace('1200 midnight', '0000')
df = df.astype({'shift': int})
df['shift'] = df['shift']/100

df.loc[df['shift'] == 99, 'time'] = np.nan
```
